```python
null_values=review.isnull().sum()
print(null_values)
```

```python
year=review.columns[4:]
review_year=review[year]
```

```python
review_year.head(5)
```

Now only Year Colums are shown now we have already identified the number of missing values

we will fill the median values in the null places as there are not many rows that are empty

```python
year = '2020'
print("Minimum value:", review_year[year].min())
print("Median value :", review_year[year].median())
print("Mean value   :", review_year[year].mean())
print("Maximum value:", review_year[year].max())
```

the data is from very small values to very large values so thats why we have to use median instead of mode or mean.
mean is biased due to extreme values.
missing values is less than 10% or around 6 to 7% so we can proceed and add that missing values instead f removing any row or column

```python
# Filling missing values using median for each year
review_year_filled = review_year.fillna(review_year.median())
```

```python
# Checking again for any other missing values
print(review_year_filled.isnull().sum())
```

now my data has no missing values all values are filled now.

i am going to combine my this clean dataset to my original dataset.

```python
review_cleaned = review.copy()

year_cols = review.columns[4:]
review_cleaned[year_cols] = review_year_filled.values 

print(review_cleaned.head())
print(review_cleaned.isnull().sum())

```

```python
#saving the cleaned data
review_cleaned.to_csv('review_cleaned.csv', index=False)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

review_cleaned = pd.read_csv('review_cleaned.csv')

# List of countries to analyze
countries = ['United States','China','India','Russian Federation','France',
             'United Kingdom','Japan','Brazil','Australia','Germany']

# Filter dataset
df_selected = review_cleaned[review_cleaned['Country Name'].isin(countries)]

# Years columns
years = [str(y) for y in range(1990, 2024)]
```

now historical gdp plot per capita trends

```python
plt.figure(figsize=(12,6))
for index, row in df_selected.iterrows():
    plt.plot(range(1990,2024), row[years], label=row['Country Name'])

plt.title('GDP per Capita Trends (1990–2024)')
plt.xlabel('Year')
plt.ylabel('GDP per Capita (PPP $)')
plt.legend()
plt.grid(True)
plt.show()
```

now going to show the data for the next 5 year from 2025-2029

```python
plt.figure(figsize=(12,6))

years_np = np.array(range(1990,2024)).reshape(-1,1)
forecast_years = np.array(range(2024,2030)).reshape(-1,1)
all_years = np.concatenate([years_np, forecast_years])

for index, row in df_selected.iterrows():
    gdp = row[years].values
    
    poly = PolynomialFeatures(degree=3)
    years_poly = poly.fit_transform(years_np)
    model = LinearRegression()
    model.fit(years_poly, gdp)
    

    all_years_poly = poly.transform(all_years)
    pred_all = model.predict(all_years_poly)
    

    plt.plot(range(1990,2024), pred_all[:len(years_np)], color='blue', alpha=0.6)
    

    plt.plot(range(2024,2030), pred_all[len(years_np):], color='red', linestyle='--', alpha=0.7)
    

    plt.text(2029, pred_all[-1], row['Country Name'], fontsize=9)

plt.title('GDP per Capita Forecast (2025–2030) for Selected Countries')
plt.xlabel('Year')
plt.ylabel('GDP per Capita (PPP $)')
plt.grid(True)
plt.show()
```

Full visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load the data
df = pd.read_csv('review_cleaned.csv')

# Get year columns (1990-2024)
year_cols = [str(year) for year in range(1990, 2025)]

# Prepare data for visualization
countries = df['Country Name'].values
data_matrix = df[year_cols].values

print(f"Total countries: {len(countries)}")
print(f"Years covered: {year_cols[0]} to {year_cols[-1]}")
```

```python
# VISUALIZATION 2: Sorted Heatmap (by 2024 GDP)
plt.figure(figsize=(20, 30))

# Sort countries by 2024 GDP
sort_idx = np.argsort(df['2024'].values)[::-1]
data_sorted = data_matrix[sort_idx]
countries_sorted = countries[sort_idx]
data_log_sorted = np.log10(data_sorted + 1)

sns.heatmap(data_log_sorted, 
            cmap='RdYlGn',
            xticklabels=[year_cols[i] if i % 5 == 0 else '' for i in range(len(year_cols))],
            yticklabels=countries_sorted,
            cbar_kws={'label': 'log10(GDP per capita PPP)'},
            linewidths=0)

plt.title('GDP per Capita Trends - Sorted by 2024 GDP (High to Low)\nLog Scale Heatmap', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Country (Sorted by 2024 GDP)', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('heatmap_sorted_by_gdp.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Regional Comparison (Top 20 by 2024 GDP)
plt.figure(figsize=(18, 10))

# Get top 20 countries by 2024 GDP
top_20_idx = np.argsort(df['2024'].values)[-20:][::-1]
top_20_countries = countries[top_20_idx]
top_20_data = data_matrix[top_20_idx]

for i, country in enumerate(top_20_countries):
    plt.plot(years, top_20_data[i], label=country, linewidth=2, marker='o', markersize=3)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
plt.grid(True, alpha=0.3)
plt.title('GDP per Capita Trends - Top 20 Countries by 2024 GDP', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('GDP per Capita PPP', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('top_20_countries.png', dpi=150, bbox_inches='tight')
plt.show()
```
